# SQL for WCD Open League Stats

```python
import pandas as pd
import numpy as np
from pandasql import sqldf
```

```python
pysqldf = lambda q: sqldf(q, globals())
```

```python
wcd = pd.read_csv('WCDSQL.csv')
```

```python
# Separate Flights Templar
FT = wcd['Team'] != 'Flights Templar'
FT.value_counts()
WCD = wcd.mask(FT)
WCD = WCD.dropna(thresh=12)
```

```python
# Count Legs by player on Flights Templar
query = """
        SELECT Player, COUNT(Game)
        FROM WCD
        GROUP BY Player
        """

result = pysqldf(query)
result
```

```python
# Players having over 100 Legs
query = """
        SELECT Player, COUNT(Game)
        FROM WCD
        GROUP BY Player
        HAVING COUNT(Game) > 100
        """

result = pysqldf(query)
result
```

```python
# Players with over 100 Cricket Legs
query = """
        SELECT Player, COUNT(Game)
        FROM WCD
        WHERE Game LIKE 'Crick%'
        GROUP BY Player
        HAVING COUNT(Game) > 100
        """

result = pysqldf(query)
result
```

```python
# 301 Averages for players with over 75 legs
query = """
        SELECT Player, COUNT(Game), AVG(TDA)
        FROM WCD
        WHERE Game LIKE '301%'
        GROUP BY Player
        HAVING COUNT(Game) > 75
        """

result = pysqldf(query)
result
```

```python
# And again for the whole league for players with over 100 legs, with averages
query = """
        SELECT Player, COUNT(Game), AVG(TDA) AS '3_Dart_Average_301'
        FROM wcd
        WHERE Game LIKE '301%'
        GROUP BY Player
        HAVING COUNT(Game) > 100
        """

result = pysqldf(query)
result
```
